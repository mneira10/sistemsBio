```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
def fx(y):
    return b/(1+y**h)

def fy(x):
    return c/(1+x**n)
def fx2(y):
    return (c/y -1 )**(1/n)

def fopt(y):
    return fx(y)-fx2(y)
```

```python
b = 3
c = 3
n = 2
h = 2

maxy = maxx = 3.3
npoints=100
xs = np.linspace(0,maxx,npoints)
y = fy(xs)
ys = np.linspace(0,maxy,npoints)
x = fx(ys)
```

```python
root1 = optimize.newton(fopt,0.001)
root2 = optimize.newton(fopt,1)
root3 = optimize.newton(fopt,2.96)
roots = [root1,root2,root3]
roots = np.array(roots)
print(roots)
print(fy(roots))
```

```python
plt.plot(xs,y,linewidth=2,label=r'$y=\frac{c}{1+x^n}$')
plt.plot(x,ys,linewidth=2,label=r'$x=\frac{b}{1+y^h}$')
for root in roots:
    plt.plot([root,root],[min(ys),max(ys)],c='black',linestyle='--',linewidth=1)
    plt.plot([min(xs),max(xs)],[fy(root),fy(root)],c='black',linestyle='--',linewidth=1)
plt.legend(loc=7)
plt.title("b=c=3, h=n=2.5")
plt.savefig('1b.png')
```
