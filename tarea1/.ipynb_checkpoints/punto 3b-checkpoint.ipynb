```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
def f(x,v1,v0):
    return (v0+v1*x**2)/(1+x**2)

def g(x):
    return x
```

```python
#ejemplo iii
v1=2
v0=0.1
xs = np.linspace(0,1.5,1000)
fs = f(xs,v1,v0)
gs = g(xs)

coefs = [1,-v1,1,-v0]
sols = list(np.roots(coefs))



ys = np.linspace(0,max(max(fs),max(gs)),100)

for i in sols:
    plt.plot([i]*len(ys),ys,c='black',linestyle = '--')

    
sols = [max(xs)] + sols + [0]

print(sols)
ax = plt.axes()

for i in range(len(sols)-1): 
    fourth = (sols[i] - sols[i+1])/4
    ymid = max(ys)/2
    right = sols[i]-fourth
    left = right-fourth*2
    if(f(left,v1,v0)>g(left)):
        ax.arrow(left, ymid, fourth*2, 0, head_width=0.05, head_length=0.02, fc='k', ec='k')
    else:
        ax.arrow(right, ymid, -fourth*2, 0, head_width=0.05, head_length=0.02, fc='k', ec='k')
#     break



# ax.arrow(0, 0, 0.5, 0.5, head_width=0.05, head_length=0.1, fc='k', ec='k')




plt.plot(xs,fs,label="f(x)")
plt.plot(xs,gs,label="g(x)")
plt.legend()
plt.xlabel("x")
plt.title(r"$v_1 = {:.2f}\ \  v_0 = {:.2f}$".format(v1,v0))
plt.savefig("iii.pdf")
```

```python
#ejemplo v
v1=4
v0=1
xs = np.linspace(0,5,1000)
fs = f(xs,v1,v0)
gs = g(xs)


coefs = [1,-v1,1,-v0]
sols = (np.roots(coefs))
sols = np.real(sols[np.isreal(sols)])
sols = list(sols)




ys = np.linspace(0,max(max(fs),max(gs)),100)

for i in sols:
    plt.plot([i]*len(ys),ys,c='black',linestyle = '--')

    
sols = [max(xs)] + sols + [0]


ax = plt.axes()

for i in range(len(sols)-1): 
    fourth = (sols[i] - sols[i+1])/4
    ymid = max(ys)/2
    right = sols[i]-fourth
    left = right-fourth*2
    if(f(left,v1,v0)>g(left)):
        ax.arrow(left, ymid, fourth*2, 0, head_width=0.2, head_length=0.1, fc='k', ec='k')
    else:
        ax.arrow(right, ymid, -fourth*2, 0, head_width=0.2, head_length=0.1, fc='k', ec='k')


plt.plot(xs,fs,label="f(x)")
plt.plot(xs,gs,label="g(x)")
plt.legend()
plt.xlabel("x")
plt.title(r"$v_1 = {:.2f}\ \  v_0 = {:.2f}$".format(v1,v0))
plt.savefig("v.pdf")
```
