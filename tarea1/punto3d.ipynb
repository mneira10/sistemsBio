```python
import numpy as np
import matplotlib.pyplot as plt
import itertools as itera
```

```python
def numSols(v0,v1):
    coefs = [1,-v1,1,-v0]
    sols = (np.roots(coefs))
    return sum(np.isreal(sols))
```

```python
numSols(1,4)
```

```python
v0 = np.linspace(0,0.2,10)
v1 = np.linspace(0,30,10)

for i in itera.product(v0,v1):
    color='black'
    if(numSols(i[0],i[1])>1):
        color = 'red'
        
    plt.scatter(i[0],i[1],c=color)
    
plt.xlabel('v0')
plt.ylabel('v1')
```

```python
def vo(v1):
    return ((1/3)*(v1 +(v1**2-3)**0.5))**2

def a(v1):
    return (1/3)*(v1 +(v1**2-3)**0.5)
def al(v1):
    return (1/3)*(v1 -(v1**2-3)**0.5)

def v0(v1):
    return a(v1)**2 *(v1-2*a(v1))
def v02(v1):
    return al(v1)**2 *(v1-2*al(v1))
```

```python
v1 = np.linspace(3**0.5+0.0000000001,100,1000)
# v12 = np.linspace(3**0.5+0.0000000001,100,1000)
v0s = v0(v1)
v0s1 = v02(v1)
```

```python
fig, ax = plt.subplots()

# ax.fill_between(v1,v0s1,v0s,label = 'bla')
plt.xlim(0,0.2)
plt.ylim(0,10)
# ax.invert_xaxis()

plt.plot(v0s,v1,label=r'$v_0=a^2(v_1-2a);\ \ a=\frac{1}{3}(v1-\sqrt{v1^2-3})$')
plt.plot(v0s1,v1,label=r'$v_0=a^2(v_1-2a);\ \ a=\frac{1}{3}(v1+\sqrt{v1^2-3})$')
plt.xlabel
# plt.scatter(0.1,2)
# plt.scatter(1,4,c='red')
plt.scatter(0.2,1,c='red')
plt.xlabel(r'$v_0$')
plt.ylabel(r'$v_1$')
plt.legend()
plt.savefig("curvas.pdf")
```
