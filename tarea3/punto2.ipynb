```python
import numpy as np
import matplotlib.pyplot as plt
import math as math
from scipy import integrate
```

```python
def poisson(xs,lamb):
    return lamb**xs * np.exp(-lamb)/math.factorial(xs)
def pn(k,g,n):
    return ((k+g*n)/k)**(4*k/g)*np.exp(-2*n)
```

```python
xs=np.array([i for i in range(21)])
```

```python
lambs = [0.5,4,7]
```

```python
fig = plt.figure(figsize=(8,20))
for i,l in enumerate(lambs):
    plt.subplot(3, 1, i+1)
    pys = []
    for x in xs:
        pys.append(poisson(x,l))
    plt.plot(xs,pys,label='Poisson')
    g=1/5
    fp = pn(l*g,g,xs)
#     print(len(xs))
#     print(len(fp))
    fp /= integrate.simps(fp,xs)
    plt.title(r'$\lambda = {}$'.format(l))
    plt.plot(xs,fp,label='Fokker-Planck')
    plt.xlabel('n')
    plt.ylabel('p(n)')
    plt.legend()
plt.savefig('punto2.png')
```
