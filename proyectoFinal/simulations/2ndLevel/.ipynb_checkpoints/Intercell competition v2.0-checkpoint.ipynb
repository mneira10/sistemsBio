```python
import numpy as np 
import matplotlib.pyplot as plt 
```

### Global parameters

```python
h  = 2
gamma = 1.0/30
beta = 1.0/10
#normal rate of growth, about onces every 20 mins
phi = 1.0/20.0
# penalty for having plasmids, small number, eyeballed
c = 0.01
# eyeballed parameter, kill them every ? mins
gammaB = 1/20000

```

```python
print(bss(nss(30)))
```

### Equations

#### Plasmids

We'll keep the equations for plasmids

```python
def alpha(n1,n2,ki):
    return beta/(1+((n1+n2)/ki)**h)
def nplus(n1,n2,ki):
    return n1*alpha(n1,n2,ki)
def nminus(n1,n2,ki):
    return n1*gamma
```

#### Bacteria

Well define the behaviour of bacteria in the following way:

$$\dot{B_i} = r_i(n_i) - \gamma_B \sum_i B_i$$

$$r_i(n_i) = \Phi(1-c\cdot n_i)$$

```python
def r(ni):
    return phi*(1-c*ni)
def bplus(ni):
    return r(ni)
def bminus(b1,b2):
    return gammaB*(b1+b2)
```

We can also find the steady state for when there is only one type of cell:

$$\dot{B_i} = 0 \Rightarrow r_i(n_i) = \gamma_B B_i$$

$$ \Phi(1-cn_i) = \gamma_B B_i$$

$$B_{ss} = \frac{\Phi(1-cn_{ss})}{\gamma_B} $$

```python
def nss(k):
    return k*(beta/gamma -1 )**(1/h)
def bss(nss):
    return phi*(1-c*nss)/gammaB
```

### Simulation

Gillespie functions

```python
def tau(s,r):
    return np.log(1/r)/s

def events(b1,b2,n1,n2,k1,k2):
    return [nplus(n1,n2,k1),nminus(n1,n2,k1),nplus(n2,n1,k2),nminus(n2,n1,k2),
            bplus(n1),bminus(b1,b2),bplus(n2),bminus(b1,b2)]

def action(b1_list,b2_list,n1_list,n2_list,events,r):
    normedEvents = np.array(events)/sum(events)
    normedEvents = np.cumsum(normedEvents)
    
    if(r<normedEvents[0]):
        n1_list.append(n1_list[-1]+1)
        n2_list.append(n2_list[-1])
        b1_list.append(b1_list[-1])
        b2_list.append(b2_list[-1])
        
    elif(r<normedEvents[1]):
        n1_list.append(n1_list[-1]-1)
        n2_list.append(n2_list[-1])
        b1_list.append(b1_list[-1])
        b2_list.append(b2_list[-1])
        
    elif(r<normedEvents[2]):
        n1_list.append(n1_list[-1])
        n2_list.append(n2_list[-1]+1)
        b1_list.append(b1_list[-1])
        b2_list.append(b2_list[-1])
        
    elif(r<normedEvents[3]):
        n1_list.append(n1_list[-1])
        n2_list.append(n2_list[-1]-1)
        b1_list.append(b1_list[-1])
        b2_list.append(b2_list[-1])
    elif(r<normedEvents[4]):
        n1_list.append(n1_list[-1])
        n2_list.append(n2_list[-1])
        b1_list.append(b1_list[-1]+1)
        b2_list.append(b2_list[-1])
    elif(r<normedEvents[5]):
        n1_list.append(n1_list[-1])
        n2_list.append(n2_list[-1])
        if(b1_list[-1]>0):
            b1_list.append(b1_list[-1]-1)
        else:
            b1_list.append(b1_list[-1])
        b2_list.append(b2_list[-1])
    elif(r<normedEvents[6]):
        n1_list.append(n1_list[-1])
        n2_list.append(n2_list[-1])
        b1_list.append(b1_list[-1])
        b2_list.append(b2_list[-1]+1)
    else:
        n1_list.append(n1_list[-1])
        n2_list.append(n2_list[-1])
        b1_list.append(b1_list[-1])
        if(b2_list[-1]>0):
            b2_list.append(b2_list[-1]-1)
        else:
            b2_list.append(b2_list[-1])

    
```

### Single cell simulation

```python
def sim1Cell(k1,k2):
    n1 = [1.0]
    n2 = [1.0]
    b1 = [5.0]
    b2 = [5.0]
    t = [0.0]

    totT = 800
    while(t[-1]<totT):


        r1= np.random.random()

        eventsNow = events(b1[-1],b2[-1],n1[-1],n2[-1],k1,k2)

        t.append(t[-1]+tau(sum(eventsNow),r1))

        r2 = np.random.random()
        action(b1,b2,n1,n2,eventsNow,r2)
    return t,n1,n2,b1,b2
    
```

### Multiple cell simulation

```python
def multiSim(k1,k2,N=400):
    n1s = []
    n2s = []
    b1s = []
    b2s = []
    ts= []


    for i in range(N):
        print('\r'+"{:.2f}".format(i*100/N),end='')
        t,n1,n2,b1,b2=sim1Cell(k1,k2)
        n1s.append(n1)
        n2s.append(n2)
        b1s.append(b1)
        b2s.append(b2)
        ts.append(t)
    return ts,n1s,n2s,b1s,b2s
```

### Multiple cell simulation and plot 

```python
def simAndPlot(ks):
    fig = plt.figure(figsize=(30,6*len(ks)))
    for w,kk in enumerate(ks):
        k1 = kk[0]
        k2 = kk[1]
        
        print(k1,k2)
        
        ts,n1s,n2s,b1s,b2s = multiSim(k1,k2)
        lastn1 = []
        lastn2=[]
        lastb1= []
        lastb2 = []
        
        
        #plasmids over time
        plt.subplot(len(ks),4,w*4+1)
        for i in range(len(ts)):
            lastn1.append(n1s[i][-1])
            lastn2.append(n2s[i][-1])
            plt.plot(ts[i],n1s[i],color='red'  ,alpha=0.2)
            plt.plot(ts[i],n2s[i],color='black',alpha=0.2)

        plt.plot(ts[0],n1s[0],color='red'  ,label = 'n1',alpha=0.2)
        plt.plot(ts[0],n2s[0],color='black',label = 'n2',alpha=0.2)
        plt.xlabel('time (mins)')
        plt.ylabel('# of plasmids')
        plt.legend()
        plt.title('k1 = {} k2 = {}'.format(k1,k2))

        
        #plasmid histograms
        plt.subplot(len(ks),4,w*4+2)

        nbins = np.linspace(0,90,30)
        plt.hist(lastn1,bins=nbins,alpha = 0.5,color = 'red'  ,label = 'n1',density=True)
        plt.hist(lastn2,bins=nbins,alpha = 0.5,color = 'black',label = 'n2',density=True)
        plt.plot([nss(k1),nss(k1)],[0,0.15],linestyle='--',color = 'red',label='n1 steady state')
        plt.plot([nss(k2),nss(k2)],[0,0.15],linestyle='--',color = 'black',label='n2 steady state')
        plt.xlabel("# of plasmids")
        plt.ylabel("frequency")
        plt.legend()
        plt.title('k1 = {} k2 = {}'.format(k1,k2))
        
    
        
        #bacteria over time
        
        plt.subplot(len(ks),4,w*4+3)
        for i in range(len(ts)):
            lastb1.append(b1s[i][-1])
            lastb2.append(b2s[i][-1])
            plt.plot(ts[i],b1s[i],color='red'  ,alpha=0.2)
            plt.plot(ts[i],b2s[i],color='black',alpha=0.2)

        plt.plot(ts[0],b1s[0],color='red'  ,label = 'b1',alpha=0.2)
        plt.plot(ts[0],b2s[0],color='black',label = 'b2',alpha=0.2)
        plt.xlabel('time (mins)')
        plt.ylabel('# of bacteria')
        plt.legend()
        plt.title('k1 = {} k2 = {}'.format(k1,k2))
        
        #bacteria histograms
        plt.subplot(len(ks),4,w*4+4)

        nbins = np.linspace(0,1000,40)
        plt.hist(lastb1,bins=nbins,alpha = 0.5,color = 'red'  ,label = 'b1',density=True)
        plt.hist(lastb2,bins=nbins,alpha = 0.5,color = 'black',label = 'b2',density=True)
        plt.plot([bss(nss(k1)),bss(nss(k1))],[0,0.15],linestyle='--',color = 'red',label='n1 steady state')
        plt.plot([bss(nss(k2)),bss(nss(k2))],[0,0.15],linestyle='--',color = 'black',label='n2 steady state')
        plt.xlabel("# of bacteria")
        plt.ylabel("frequency")
        plt.legend()
        plt.title('k1 = {} k2 = {}'.format(k1,k2))
        
        
        plt.savefig('segundaEtapa.png')
        
```

```python
simAndPlot([[10,50]])
```
