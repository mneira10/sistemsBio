```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
h  = 2
gamma = 1.0/30
beta = 1.0/10
```

```python
ns = np.linspace(0,100,1000)
```

Given that plasmids within a cell lie in the range of 30-50, a first approach for $K_i$ would be:

```python
50/(beta/gamma-1)**(1/h)
```

```python
def alpha(n,K):
    return beta/(1+(n/K)**h)
def nprime(n,K):
    return n*(alpha(n,K)-gamma)
def nss(k):
    return k*(beta/gamma - 1)**(1/h)
```

```python
ks = [10,20,30,40]
colors = ['red','blue','green','black']
for i,k in enumerate(ks):
    ys = nprime(ns,k)
    plt.plot(ns,ys,label="{}".format(k),color=colors[i])
    plt.legend()
plt.xlabel(r'$n$')
plt.ylabel(r'$\dot{n}$')
plt.ylim(0,1)
```
