```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
h  = 2
gamma = 1.0/30
beta = 5.0
```

```python
ns = np.linspace(0,100,1000)
```

```python
def alpha(n,K):
    return beta/(1+(n/K)**h)
def nprime(n,K):
    return n*(alpha(n,K)-gamma)
def nss(k):
    return k*(beta/gamma - 1)**(1/h)
```

```python
ks = [1,3,5,10]
colors = ['red','blue','green','black']
for i,k in enumerate(ks):
    nssp = nss(k)
    plt.plot([nssp,nssp],[0,25],color=colors[i])
    ys = nprime(ns,k)
    plt.plot(ns,ys,label="{}".format(k),color=colors[i])
    plt.legend()
```
