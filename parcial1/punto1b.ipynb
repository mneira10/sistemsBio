```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
vx = 0.1
vy = 0
kx = 4
ky = 2
gx = 10
G = 1/gx
```

```python
def yk(x):
    return vy+ky*x
def xk(x):
    return kx/(x-vx) -1 
    
```

```python
x = np.linspace(0.2000001,5,100)
y1 = yk(x)
y2 = xk(x)
```

```python
plt.plot(x,y1,label = 'y nullclein')
plt.plot(x,y2,label = 'x nullclein')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.savefig('sistemA.png')
```

```python
def yk(x):
    return vy+ky*x
def xk(x):
    return kx*x**2/((x-vx)*(1+x**2)) -1 
    
```

```python
x = np.linspace(0.4,0.6,100)
y1 = yk(x)
y2 = xk(x)

plt.plot(x,y1,label = 'y nullclein')
plt.plot(x,y2,label = 'x nullclein')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(0.8,1.2)
# plt.yscale('log')

plt.legend()
plt.savefig('sistemB.png')
```

```python
from scipy import optimize

vx = 0.1
vy = 0
kx = 4
ky = 2
gx = 10
G = 1/gx

def f(x):
    return vy+ky*x-kx*x**2/((x-vx)*(1+x**2)) +1 

def ynull(x):
    return vy+ky*x

root = optimize.newton(f, 0.4)
print("xss = {:.2f}, yss = {:.2f}".format(root,ynull(root)))
```

```python
x = np.linspace(0.4,0.6,100)
y1 = yk(x)
y2 = xk(x)

plt.plot(x,y1,label = 'y nullclein')
plt.plot(x,y2,label = 'x nullclein')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(0.8,1.2)
plt.plot([root,root],[0,2],linestyle ="--" ,c='black' ,label = r'$x_{ss}$')
plt.plot([0.4,0.6],[ynull(root),ynull(root)],linestyle ="--" ,c='red' ,label = r'$y_{ss}$')
# plt.yscale('log')

plt.legend()
plt.savefig('stablePoint.png')
```
